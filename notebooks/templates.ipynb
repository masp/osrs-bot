```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('../snapshots/play.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(image)
plt.axis('on')
plt.show()
```

```python
template = cv2.imread('../templates/play_button.png')
template = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(template)
plt.axis('on')
plt.show()
```

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
template = cv2.cvtColor(template, cv2.COLOR_RGB2GRAY)

mat = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)
```

```python
loc = np.where(mat > 0.9)
w, h = template.shape[::-1]


plt.figure(figsize=(10,10))
plt.imshow(image[loc[0][0]:loc[0][0]+h, loc[1][0]:loc[1][0]+w])
plt.axis('on')
plt.show()
```
